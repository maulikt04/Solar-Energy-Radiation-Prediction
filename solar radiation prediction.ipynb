```python
s_r_1 = s_r.drop(columns = ['UNIXTime', 'Data', 'Time', 'WindDirection(Degrees)', 'Speed', 'TimeSunRise', 'TimeSunSet'])
s_r_1
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
s_r_1[["ScaledTemp", "ScaledPressure", "ScaledHumidity", "ScaledRadiation"]] = scaler.fit_transform(
    s_r_1[["Temperature", "Pressure", "Humidity", "Radiation"]])
s_r_2 = s_r_1.drop(columns = ["Temperature", "Pressure", "Humidity", "Radiation"])
s_r_2
```

```python
#Setting the value for X and Y
x = s_r_2.drop('ScaledRadiation', axis = 'columns')
y = s_r_2['ScaledRadiation']
```

```python
#Splitting the dataset
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 150)
```

# Multiple Linear Regression

```python
#Fitting the Multiple Linear Regression model
from sklearn.linear_model import LinearRegression
mlr = LinearRegression()  
mlr.fit(x_train, y_train)
```

```python
print("Intercept: ", mlr.intercept_)
print("Coefficients:")
list(zip(x, mlr.coef_))
```

```python
#Prediction of test set
y_pred_mlr= mlr.predict(x_test)
#Predicted values
print("Prediction for test set: {}".format(y_pred_mlr))
```

```python
#Actual value and the predicted value
mlr_diff = pd.DataFrame({'Actual value': y_test, 'Predicted value': y_pred_mlr})
mlr_diff.head()
```

```python
#Model Evaluation
from sklearn import metrics
meanAbErr = metrics.mean_absolute_error(y_test, y_pred_mlr)
meanSqErr = metrics.mean_squared_error(y_test, y_pred_mlr)
rootMeanSqErr = np.sqrt(metrics.mean_squared_error(y_test, y_pred_mlr))
print('R squared: {:.2f}'.format(mlr.score(x,y)*100))
print('Mean Absolute Error:', meanAbErr)
print('Mean Square Error:', meanSqErr)
print('Root Mean Square Error:', rootMeanSqErr)
```

```python
from sklearn.model_selection import KFold
model = LinearRegression()
kfold_validation = KFold(10)

from sklearn.model_selection import cross_val_score
result = cross_val_score(model, x, y, cv = kfold_validation)

print(result*100)
print(np.mean(result*100))
```

```python
#Testing score by cross validation, here I take number of K-Fold is 25 for testing:
test_scores = cross_val_score(model, x_test, y_test , cv = 25) 
test_scores*100
```

```python
#Mean of testing score:
print(np.mean(test_scores*100))
```

```python
import tensorflow as tf
ann = tf.keras.models.Sequential()
```

```python
#Add input and first hidden layer:
ann.add(tf.keras.layers.Dense(units=5, input_dim=3,  kernel_initializer='normal', activation = 'relu'))
```

```python
#Add second hidden layer:
ann.add(tf.keras.layers.Dense(units=3,kernel_initializer='normal', activation = 'relu'))
```

```python
#Add output layer:
ann.add(tf.keras.layers.Dense(units=1,kernel_initializer='normal', activation = 'sigmoid'))
```

```python
#compiling the ANN:
ann.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
#Training the ANN on the Training Set:
ann.fit(x_train, y_train, batch_size = 20, epochs = 10)
```
